```python
from nltk.tokenize import sent_tokenize
from bs4 import BeautifulSoup as soup
import requests as req
import os
```

```python
#paths of directories
path = os.path.join(os.getcwd() ,'data')
dataDirPath = os.path.join(path , 'dataDir')
metaDirPath = os.path.join(path , 'mataDir' )

try:
    os.mkdir('data')
    os.makedirs(dataDirPath)
    os.makedirs(metaDirPath)
except:
    print('Dir already exits')
```

```python
'''
extract the data from the HTML from each review card and makes a list of
'title + comment' and its 'meta data' and append it in reviewsData list.

'''
def parseReviews(cards):
    reviewsData = []
    for card in cards:
        try:
            dateOfStay = card.find('span', {'class':"_34Xs-BQm"}).text.strip()
        except:
            dateOfStay = ''
        try:
            noOfReviewsContrib = card.find('span', {'class':"_1fk70GUn"}).text.strip()
        except:
            noOfReviewsContrib = ''
        try:
            native = card.find('span', {'class':"default _3J15flPT small"}).text.strip()
        except:
            native = ''
        try:
            reviewsHelpful = card.find('span', {'class':"_1fk70GUn"}).text.strip()
        except:
            reviewsHelpful = ''
        try:
            title = card.find('a', {'class':"ocfR3SKN"}).text.strip()+'. '
        except:
            title = ''
        try:
            comment = card.find('q', {'class':"IRsGHoPm"}).text.strip()
            if(comment[-1] == 'â€¦'):
                comment = sent_tokenize(comment)
                comment = ' '.join(comment[0:-1])
        except:
            comment = ''
        meta = f' {dateOfStay}.\n no of reviews contributed : {noOfReviewsContrib}.\n home : {native}.\n review found helpful : {reviewsHelpful}'
        reviewsData.append([title+comment , meta])
    return reviewsData
```

```python
def extractReviews(hotelUrl):
#     generalize urls 
    preUrlPart, postUrlPart = hotelUrl.split('Reviews')
    preUrlPart += 'Reviews-or'
    postUrlPart += '#REVIEWS'
    
    
#     scraping page from the net
    try:
        r = req.get(hotelUrl)
    except:
        print("Entered Url is not correct")
        return
    
    htmlPage = soup(r.content ,'html.parser')
    
#     collectsall review card if any on the entered url page
    reviewCards = htmlPage.find('div', {'class' : '_2wrUUKlw _3hFEdNs8'})
    
#     if 
    if reviewCards == None:
        print("No reviews found")
        return

#     noOfPages = No of pages from which we need to scrape comments
    noOfPages = htmlPage.find_all('a', {'class' : 'pageNum'})
        
    if len(noOfPages) == 0:
        noOfPages = 1
    else:
        noOfPages = int(noOfPages[-1].text.strip())
        print("No of pages", noOfPages)
        
#     reviewNo = Current review no
    currentPageNo = 1
    reviewNo = 1
        
    while(currentPageNo <= noOfPages):
        print("loading ..." ,currentPageNo)
        reviewCards = htmlPage.find_all('div', {'class' : '_2wrUUKlw _3hFEdNs8'})       
        
        reviewsData = parseReviews(reviewCards)
        
        for reviewData in reviewsData:
#     stores the comment and comment_meta data in seperate file for each review
            with open(os.path.join(dataDirPath, f'file{reviewNo}.txt'), 'w', encoding="utf-8") as file:
                file.write(reviewData[0])
            with open(os.path.join(metaDirPath, f'file{reviewNo}.txt'), 'w', encoding="utf-8") as file:
                file.write(reviewData[1])
                file.write('\n'+reviewData[0])
            reviewNo = reviewNo+1
        
#      new url for comments on another page
        Url = f"{preUrlPart}{currentPageNo * 5}{postUrlPart}"
        currentPageNo = currentPageNo+1
        try:
            r = req.get(Url)
            htmlPage = soup(r.content ,'html.parser')
        except:
            print(rf"Error:: could not scrape this page.\n Total pages scraped :: {currentPageNo}")
    
    print("Scrapping completed")
```
